```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/sample_students.csv')
```

Distribution of Exam Scores

```python
plt.figure(figsize=(6,4))
sns.histplot(df["exam_score"], bins=5, kde=True)
plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()
```

```python

# 2. Attendance vs Exam Score (scatter)
plt.figure(figsize=(6,4))
sns.scatterplot(x="attendance_rate", y="exam_score", hue="performance_band", data=df, s=100)
plt.title("Attendance Rate vs Exam Score")
plt.show()

```

```python
# 3. Correlation Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```

```python

# 4. Boxplot of Exam Score by Term
plt.figure(figsize=(6,4))
sns.boxplot(x="term", y="exam_score", data=df)
plt.title("Exam Scores Across Terms")
plt.show()
```

```python

# 5. Average Exam Score by Performance Band
plt.figure(figsize=(6,4))
sns.barplot(x="performance_band", y="exam_score", data=df, estimator="mean")
plt.title("Average Exam Score by Performance Band")
plt.show()
```
